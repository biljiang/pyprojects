```python
import matplotlib.pylab as plt 
from sklearn.preprocessing import StandardScaler 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import classification_report 
```

```python
df=df.drop('manufacturer',axis=1)
df=df.dropna()
```

```python
#df.inst_date.apply(lambda x: datetime.strptime(str(x),"%Y-%m-%d"))
#df.inst_date.apply(lambda x:x.year)
df["inst_date"]=df.inst_date.apply(lambda x: x.year)
```

```python
df["inst_date"]=df.inst_date.apply(str)
df["rs485_route_qty"]=df.rs485_route_qty.apply(str)
```

```python
df_dummies = pd.get_dummies(df[['wiring_mode', 'meter_mfr', 'd_meter_type',
       'meter_model', 'tmnl_mfr', 'tmnl_model', 'inst_date',
       'comm_mode', 'comm_prot_code', 'rs485_route_qty']])
```

```python
df_dummies.columns
```

```python
Y=1-df.pap_r_flag.apply(int)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(df_dummies,Y,test_size=0.3,random_state=0)
```

```python
#classifier = LogisticRegression(random_state=4,class_weight="balanced") 
classifier = LogisticRegression(random_state=4) 
classifier.fit(X_train, Y_train)
```

```python
Y_prob = classifier.predict_proba(X_test)
```
