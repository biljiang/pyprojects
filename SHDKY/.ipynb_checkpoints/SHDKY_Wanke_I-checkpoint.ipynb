```python
import numpy as np
import pandas as pd
tr_data=pd.read_csv('Iavg.csv',index_col=0)
#tr_data=[tr_data!=0]
for x in tr_data:
    if tr_data[x].mean()==0:
        tr_data=tr_data.drop(columns=x)
shuffler= np.random.permutation(len(tr_data))
tr_data1=tr_data.take(shuffler)
tr1=(tr_data1-tr_data1.describe().loc['mean'])/tr_data1.describe().loc['std']
tr=tr1.iloc[:-500,:]
te=tr1.iloc[-500:,:]
```

```python
###### network from keras for SHDKY data simulation ###########
from keras.models import Sequential
from keras.layers import Dense, Activation,Input
import keras
import numpy as np


model = Sequential()

model.add(Dense(11, input_dim=11,init="normal"))
model.add(Activation('tanh'))
model.add(Dense(6, activation='tanh',init="normal"))
model.add(Dense(1, activation='linear',init="normal"))
```

```python
model.compile(loss='mean_squared_error', 
              optimizer=keras.optimizers.SGD(lr=0.2))
```

```python
Y=np.array(tr.iloc[:,5:6])
X=np.array(tr.iloc[:,0:5].join(tr.iloc[:,6:12])) 
 
```

```python
model.fit(X, Y, epochs=30,batch_size=10,
          shuffle=True,verbose=2,validation_split=0.2)  
```

```python
model.compile(loss='mean_squared_error', 
              optimizer=keras.optimizers.SGD(lr=0.02))
```
